```python
import numpy as np
```

```python
arr = np.zeros((4096, 4096, 3))
for r in range(256):
    print(r-r%16)
    for g in range(256):
        for b in range(256):
            arr[g*16+r%16][b*16+(r-r%16)//16] = (r, g, b)
```

```python
from PIL import Image
img = Image.fromarray(arr.astype(np.uint8))
img.save('bake.bmp')
```

```python
with Image.open('bake_out.bmp') as f:
    pixel_map = f.load()
```
