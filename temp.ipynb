```python
import numpy as np
```

```python
arr = np.zeros((256, 65536, 3))
for r in range(256):
    print(r)
    for g in range(256):
        for b in range(256):
            arr[r][g*256+b] = (r, g, b)
```

```python
from PIL import Image
img = Image.fromarray(arr.astype(np.uint8))
img.save('bake.bmp')
```

```python
with Image.open('bake_out.bmp') as f:
    pixel_map = f.load()
```
